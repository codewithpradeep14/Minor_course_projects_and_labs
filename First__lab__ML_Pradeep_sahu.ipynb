```python
#Question 01
def reverse_and_concatenate(A, B):
    reversed_A = A[::-1]
    result = reversed_A + B
    return result
string_A = "THANK"
string_B = " YOU!"
final_string = reverse_and_concatenate(string_A, string_B)
print(final_string)
```

```python
#Question 02
def calculate_highest_paid_female_worker(work_hours):
    highest_hours = 0
    highest_paid_female = None

    for name, hours, gender in work_hours:
        if gender == 'F' and hours > highest_hours:
            highest_hours = hours
            highest_paid_female = name

    if highest_paid_female:
        pay = highest_hours * 20
        print(f"The female worker who worked the most hours is {highest_paid_female}.")
        print(f"She should be paid ${pay}.")
    else:
        print("There are no female workers in the list.")
work_hours = [('Abby', 100, 'F'), ('Billy', 400, 'M'), ('Cassie', 800, 'F'), ('Maggi', 600, 'F'), ('Alex', 500, 'M'), ('Raj', 225, 'M'), ('Penny', 920, 'F'), ('Ben', 300, 'M')]
calculate_highest_paid_female_worker(work_hours)
```
